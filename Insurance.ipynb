```python
a=['sex', 'smoker', 'children', 'region'] #Categporical Variables
b=['age', 'bmi', 'charges'] #Contineous variables
```

```python
#Univarient for categorical variables:-
for i in a:
    print(df[i].value_counts())
    sns.countplot(x=i,data=df)
    plt.show()
```

# Analysis

```python
#Univarient for contineous variables:-
for i in b:
    sns.distplot(df[i],kde=False,bins=10,color='r')
    plt.show()
```

# Analysis

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder
df['sex']=LabelEncoder().fit_transform(df['sex'])
df['smoker']=LabelEncoder().fit_transform(df['smoker'])
df['region']=LabelEncoder().fit_transform(df['region'])
```

```python
df.head
```

```python
#Bivarient for contineous and categorical variables:-
for i in a:
    sns.swarmplot(x=i,y='charges',data=df)
    plt.show()
    sns.boxplot(x=i,y='charges',data=df)
    plt.show()
    sns.distplot(df[i],kde=False,bins=10,color='r')
    plt.show()
```

```python
#Bivarient for contineous and contineous variables 
for i in b:
    sns.scatterplot(x=i,y='charges',hue='smoker',data=df)
    plt.show()
    #smoker 1(yes) and 0(No)#
```
